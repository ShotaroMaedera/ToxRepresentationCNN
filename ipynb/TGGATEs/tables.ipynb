```python
import pandas as pd
```

```python
train_val = pd.read_csv("../../data/TGGATEs/processed/train_val_for_model_training.csv")
test = pd.read_csv("../../data/TGGATEs/processed/test_for_finding_types.csv")
```

```python
ft_list = list(train_val.columns[3:11])
ft_list
```

```python
df = pd.DataFrame()
train = train_val[train_val["fold"]!=2]
df["positive"] = (train[ft_list]>=0.5).sum()
df["negative"] = (train[ft_list]<0.5).sum()
df["ratio"] = (train[ft_list]>=0.5).mean()
display(df)

df = pd.DataFrame()
val = train_val[train_val["fold"]==2]
df["positive"] = (val[ft_list]>=0.5).sum()
df["negative"] = (val[ft_list]<0.5).sum()
df["ratio"] = (val[ft_list]>=0.5).mean()
display(df)
df = pd.DataFrame()
df["positive"] = (test[ft_list]>=0.5).sum()
df["negative"] = (test[ft_list]<0.5).sum()
df["ratio"] = (test[ft_list]>=0.5).mean()
display(df)
```
